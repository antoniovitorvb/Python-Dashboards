# Monte Carlo Simulation Method

### 1st example:

calculate the area of a circle with radius = 1 circumscribed in a square of length = 2

```python
import random as rd
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000000
circ = 0

for i in range(N):
    x = rd.uniform(-1, 1)
    y = rd.uniform(-1, 1)

    if (math.sqrt(x**2 + y**2) < 1):
        circ += 1

print('PI / 4 =', round(np.pi/4, 4))
print('circ / N =', round(circ/N, 4))
```

Testing final approximation values incrementing N

```python
N = 20000

MC = list()

for n in range(N):

    circ = 0

    for i in range(n+1):
        x = rd.uniform(-1, 1)
        y = rd.uniform(-1, 1)

        if (math.sqrt(x**2 + y**2) < 1):
            circ += 1

    # print('Exact valeu os PI / 4 =', round(np.pi/4, 4))
    # print('circ / N =', round(circ/N, 4))

    MC.append(circ/(n+1))

plt.plot(MC)
plt.axhline(np.pi/4, color = 'r')
plt.text(x = N/2, y = 0.85, s = 'Pi / 4 = {}'.format(round(np.pi/4, 4)), color = 'r', fontsize = 12)
plt.text(x = 0.7*N, y = 0.6, s = 'Final aprox.: {}'.format(MC[-1]))
plt.show()
```

### 2nd example:

Playing a game with probablitity of win P(win) = p. The game stop at 2 losses in a roll. What is the estimate of rounds played?

```python
import random as rd
import statistics as stat
import matplotlib.pyplot as plt
```

```python
rounds = list()
N = 200000
p = 0.5

for n in range(1, N+1):
    # print(n)

    r = 0 # number of rounds played
    nloss = 0 # number of losses

    while (nloss != 2):
        r += 1
        if (rd.random() < p):
            nloss = 0
        else:
            nloss += 1
    
    rounds.append(r)

y_hist = plt.hist(rounds, bins = max(rounds))[0]
plt.text(x = max(rounds)/2, y = max(y_hist)/2, s = 'Avg. Rounds = {}'.format(round(stat.mean(rounds), 3)))
plt.show()
```
