## Neste projeto, trabalharemos com dois arquivos csvs separados, contendo informações sobre o preço da gasolina no Brasil. 

### 1. Carregue os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combine-os em um único DataFrame.

```python
gas2000 = pd.read_csv('datasets\gasolina_2000+.csv', index_col=0)
gas2010 = pd.read_csv('datasets\gasolina_2010+.csv', index_col=0)

gas = pd.concat([gas2000, gas2010],
axis=0)

gas
```

### 2.Investigue as colunas e entenda o conjunto de dados usando o head() e info()

```python
gas.info()
```

### 3. Selecione a terceira entrada da coluna DATA INICIAL e verifique seu tipo.

```python
print(gas['DATA INICIAL'][2])
type(gas['DATA INICIAL'][2])
```

### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como string. Utilizando o método pd.to_datetime(), converta ambas colunas para Timestamp / Datetime.

```python
gas['DATA INICIAL'] = pd.to_datetime(gas['DATA INICIAL'])
gas['DATA FINAL'] = pd.to_datetime(gas['DATA FINAL'])
```

### 5. Crie uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência. 

```python
print(pd.DatetimeIndex(gas['DATA FINAL']).year)
pd.DatetimeIndex(gas['DATA FINAL']).month
```

```python
# gas['DATA FINAL'].apply(lambda x: '{}'.format(x.year)) + gas['DATA FINAL'].apply(lambda x: '-{}'.format(x.month))

gas['ano-mes'] = gas['DATA FINAL'].apply(lambda x: '{}'.format(x.year)) + gas['DATA FINAL'].apply(lambda x: '-{:02d}'.format(x.month))
gas
```

### 6. Utilizando o value_counts(), liste todos os tipos de produtos contidos na base de dados.

```python
gas['PRODUTO'].value_counts()
```

### 7. Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
gasolina = gas[gas['PRODUTO'] == 'GASOLINA COMUM']
gasolina
```

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

```python
# gasolina['PREÇO MÉDIO REVENDA'].mean()

# gasolina['PREÇO MÉDIO REVENDA'][gasolina['ano-mes'] == '2008-08']

'R$ '+str(round(gasolina['PREÇO MÉDIO REVENDA'][gasolina['ano-mes'] == '2008-08'].mean(), 3))
```

### 9. Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
gasolina['PREÇO MÉDIO REVENDA'][(gasolina['ano-mes'] == '2008-08') & (gasolina['ESTADO'] == 'SAO PAULO')].mean()
```

### 10. Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
gasolina[
    ['ano-mes', 'ESTADO', 'PREÇO MÁXIMO REVENDA']
][gasolina['PREÇO MÁXIMO REVENDA'] > 5].sort_values('PREÇO MÁXIMO REVENDA', ascending=False)
```

### 11. Qual a média de preço dos estados da região sul em 2012?

```python
# gasolina['DATA FINAL'].apply(lambda x: '{}'.format(x.year))

round(gasolina['PREÇO MÉDIO REVENDA']
[(gasolina['REGIÃO'] == 'SUL') & (gasolina['DATA FINAL'].apply(lambda x: '{}'.format(x.year)) == '2012')].mean(), 2)
```

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
gas_rj = gas[gas['ESTADO'] == 'RIO DE JANEIRO']
gas_rj['mes'] = gas_rj['DATA FINAL'].apply(lambda x: '{}'.format(x.month))

gas_rj = gas_rj.groupby('ano-mes')[['PREÇO MÉDIO REVENDA', 'mes']].last()
gas_rj
```

shift(n) desloca os valores do dataframe em n linhas para baixo

```python
gas_rj[gas_rj['mes'] == '12'].shift(1)
```

```python
round(gas_rj[gas_rj['mes'] == '12']['PREÇO MÉDIO REVENDA'] / gas_rj[gas_rj['mes'] == '12']['PREÇO MÉDIO REVENDA'].shift(1) - 1, 5)*100
```

Aplicando média móvel com m = 3 -> rolling(m).mean()

```python
((gas_rj[gas_rj['mes'] == '12']['PREÇO MÉDIO REVENDA'] / gas_rj[gas_rj['mes'] == '12']['PREÇO MÉDIO REVENDA'].shift(1) - 1)*100).rolling(3).mean()
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
gas2 = gas.copy()
gas2['mes'] = gas2['DATA FINAL'].apply(lambda x: '{}'.format(x.month))
```

```python
gas2 = gas2.groupby(['ano-mes', 'ESTADO'])[[
    'ano-mes', 'ESTADO',
    'PREÇO MÍNIMO REVENDA', 'PREÇO MÁXIMO REVENDA'
]].mean()

gas2.reset_index(inplace=True)
gas2
```

```python
gas2.sort_values(['ESTADO', 'ano-mes']).to_csv('gas2.csv', sep=',')
gas2
```
