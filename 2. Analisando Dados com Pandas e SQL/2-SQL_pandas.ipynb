### 2.1. Inserindo dados em uma BD

```python
import pandas as pd
import sqlite3

# instalar a extensão SQLite do alexcvzz
```

```python
conn = sqlite3.connect('web.db')
```

```python
df_data = pd.read_csv('bd_data.csv')
df_data
```

```python
df_data.index.name = 'index_name'
df_data
```

From dataframe to SQL:

```python
c = conn.cursor()
```

```python
query = 'DROP TABLE IF EXISTS data'

c.execute(query)
conn.commit()
```

```python
df_data.to_sql(name='data', con=conn, index_label='index_name')
```

### 2.2. Comandos CREATE e DROP

```python
# c = conn.cursor()
```

```python
query = '''
DROP TABLE IF EXISTS products
'''

c.execute(query)
conn.commit()
```

```python
query = '''
CREATE TABLE products(
    [product_id] INTEGER PRIMARY KEY,
    [product_name] VARCHAR(20),
    [price] NUMERIC
)
'''

c.execute(query)
conn.commit()
```

### 2.3. INSERT

```python
query = '''
INSERT INTO products values (1, 'Computer', 800),
(2, 'Printer', 300),
(3, 'Tablet', 400)
'''
c.execute(query)
conn.commit()
```

```python
df_data2 = df_data.iloc[::-2]
# df_data2

df_data2.to_sql(name='data', con=conn, if_exists='append')
# if_exists: 'fail' (RaiseError - default); 'replace' (Update); 'append' (insert)
```

Se houvesse alguma Constraint sobre o 'index' (UNIQUE ou PRIMARY KEY) daria erro!

### 2.4. SELECT e WHERE

```python
query = '''
DROP TABLE IF EXISTS data
'''

c.execute(query)
conn.commit()

df_data.to_sql(name='data', con=conn, index_label='index_name')
```

```python
query = '''
SELECT *
FROM data
'''

c.execute(query)
print('Fetch One:')
c.fetchone() # exibe apenas a primeira linha do resultado da query
```

```python
print('Fetch All:')
c.fetchall()
```

```python
df = pd.DataFrame(c.fetchall())
df
```

Retorna todas as colunas quando a coluna A for > 200:

```python
query = '''
SELECT *
FROM data
WHERE A > 200
'''

c.execute(query)
c.fetchall()
```

Melhor forma de atribuir uma query a um DataFrame:

```python
query = '''
SELECT *
FROM data
WHERE A > 200
'''

df = pd.read_sql(query, con=conn, index_col='index_name')
df
```

### 2.5. UPDATE e DELETE

```python
query = '''
UPDATE data
SET A = 300, B = 400
WHERE index_name <= 10
'''
c.execute(query)
conn.commit()
```
